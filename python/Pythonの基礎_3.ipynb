```python
import numpy as np              #Numpyの利用
#import cv2                      #OpenCVの利用　今回は利用しないためコメントアウト
import matplotlib.pyplot as plt #作図用

width = 400  #幅
height = 400 #高さ

narray = np.full( (height,width,3),0 )   #(高さ,幅,RGB),0で初期化した3次元配列

a = 2
b = 200
for y in range(height):     #yに0からheight-1までの数字が入ってループする
    for x in range(width):  #xに0からwidth-1までの数字が入ってループする
        f = y - a*x + b                  #直線の式
        if f > 0:
            narray[y,x,0] = 255 #画素のR成分のみを255にする(つまり赤色になる)
        else:
            narray[y,x,0] = 0


plt.imshow(np.uint8(narray))        #H×W×3の3次元配列を2次元カラー画像として扱う 
plt.show()                          #表示


```

```python
import numpy as np              #Numpyの利用
#import cv2                      #OpenCVの利用　今回は利用しないためコメントアウト
import matplotlib.pyplot as plt #作図用

width = 400  #幅
height = 400 #高さ

narray = np.full( (height,width,3),0 )   #(高さ,幅,RGB),0で初期化した3次元配列
cx = width/2   #中心点のX座標
cy = height/2  #中心点のY座標
rad = 200      #半径
for y in range(height):     #yに0からheight-1までの数字が入ってループする
    for x in range(width):  #xに0からwidth-1までの数字が入ってループする
        f = (y-cy)**2 + (x-cx)**2 - rad**2 #円の式
        if f < 0: 
            narray[y,x,]=255 #画素の３つの成分をを255にする(つまり白色になる)
        else:
            narray[y,x,]=0


plt.imshow(np.uint8(narray))        #H×W×3の3次元配列を2次元カラー画像として扱う  
plt.show()                          #表示
```

## 課題

以下に上記の円描画コードを示す。これを修正して短半径100長半径200の楕円形を全体が収まるように描画せよ。
楕円の向き、色などは自由とする。


```python
import numpy as np              #Numpyの利用
#import cv2                      #OpenCVの利用　今回は利用しないためコメントアウト
import matplotlib.pyplot as plt #作図用

width = 400  #幅
height = 200 #高さ

narray = np.full( (height,width,3),0 )   #(高さ,幅,RGB),0で初期化した3次元配列
cx = width/2   #中心点のX座標
cy = height/2  #中心点のY座標
rad1 = 200     #長半径
rad2 = 100      #短半径
for y in range(height):     #yに0からheight-1までの数字が入ってループする
    for x in range(width):  #xに0からwidth-1までの数字が入ってループする
        f = (y-cy)**2/(rad2)**2 + (x-cx)**2/(rad1)**2 - 1 #楕円の式
        if f < 0: 
            narray[y,x,0]=255 #画素の３つの成分をを255にする(つまり白色になる)
        else:
            narray[y,x,]=0


plt.imshow(np.uint8(narray))        #H×W×3の3次元配列を2次元カラー画像として扱う  
plt.show()                          #表示
```
