# EX_1
## 画像のグレースケール化 
### 以下のコードを完成して画像データを読み込み，グレースケール化して表示してみよう．
- グレースケール値は，⾚・緑・⻘チャンネルの平均を利⽤することとする .
$$
    I = \frac{(r + g + b)}{3}
$$

- 画像の⼊出⼒・画素値へのアクセス⽅法は雛形に書いてある.参考にしてください．



```python
# Ex1 
import numpy as np
import sys
import cv2
import matplotlib.pyplot as plt
```

```python
#入出力ファイル名指定
fname_in  = "./img_Ex1_in.jpg"
```

```python
#画像をロード, グレースケール化, float型へ変換
img = cv2.imread(fname_in)
H = img.shape[0]
W = img.shape[1]

img = np.float64(img)
```

```python
#出力画像を準備(グレースケール，float型)
img_out = np.zeros( (H,W), float )
```

```python
#フィルタ処理
for y in range(H) :
    for x in range(W) :
        #この下を編集
        img_out[y,x] = (img[y,x,0] + img[y,x,1] + img[y,x,2])/3.
        
```

```python
#display img
img = img[:, :, ::-1]  #BGR->RGB変換
plt.subplot(1, 2, 1), plt.imshow(np.uint8(img))            # 元画像貼り付け
plt.gray()                                                 #グレースケール表示
plt.subplot(1, 2, 2), plt.imshow(np.uint8(img_out))        #フィルタ画像貼り付け
plt.show()                                                 # グラフ表示
        
```

# EX2
## 画像の2値化 
### 以下のコードを完成して画像データを読み込み，グレースケール化した後，与えられた閾値により⼆値化 した結果を表示してみよう．
- 画素値が閾値以上のなら，その画素値を255にする.
- 画素値が閾値より⼩さいなら，その画素値を０とする．

```python
# Ex2.py 
import numpy as np
import sys
import cv2
import matplotlib.pyplot as plt
```

```python
#入出力ファイル名指定
fname_in  = "./img_Ex1_in.jpg"
```

```python
#しきい値設定
threshold_val = 100
threshold_val = float(threshold_val)
```

```python
#画像をロード, float型へ変換
img = cv2.imread(fname_in)
H = img.shape[0]
W = img.shape[1]

img = np.float64(img)
```

```python
#出力画像を準備(float型)
img_gray = np.zeros( (H,W), float )
img_out = np.zeros( (H,W), float )
```

```python
#グレースケール化
for y in range(H) :
    for x in range(W) :
        img_gray[y,x] = (img[y,x,0] + img[y,x,1] + img[y,x,2])/3.
        
#与えた閾値による2値化
        
for y in range(H) :
    for x in range(W) :
        #この下を編集
        if threshold_val<img_gray[y,x] :
            img_out[y,x]=255
        else:
            img_out[y,x]=0
        
```

```python
#display img
img = img[:, :, ::-1]  #BGR->RGB変換
plt.subplot(1, 2, 1), plt.imshow(np.uint8(img))            # 元画像貼り付け
plt.gray()                                                 #グレースケール表示
plt.subplot(1, 2, 2), plt.imshow(np.uint8(img_out))        #フィルタ画像貼り付け
plt.show()                                                 # グラフ表示
```

# EX3
## ⾊の変換
### 以下のコードを完成してカラー画像を読み込み，画像の⾚と⻘チャンネルを⼊れ替えた画像を表示してみよう．


```python
# Ex3.py
# 画像の赤と青チャンネルを交換する

import numpy as np
import sys
import cv2
import matplotlib.pyplot as plt
```

```python
#入出力ファイル名指定
fname_in  = "./img_Ex1_in.jpg"
```

```python
img = cv2.imread(fname_in)
img_out = np.zeros_like(img)

H   = img.shape[0]
W   = img.shape[1]
```

```python
for y in range(H) : 
    for x in    range(W) : 
        r = img[y,x,2]
        g = img[y,x,1]
        b = img[y,x,0]
        # ここを編集
        img_out[y,x,0] = r
        img_out[y,x,1] = g
        img_out[y,x,2] = b
```

```python
#display img
img = img[:, :, ::-1]  #BGR->RGB変換
img_out = img[:, :, ::-1]  #BGR->RGB変換
plt.subplot(1, 2, 1), plt.imshow(np.uint8(img))            # 元画像貼り付け
plt.subplot(1, 2, 2), plt.imshow(np.uint8(img_out))        #フィルタ画像貼り付け
plt.show()                                                 # グラフ表示
            
```
